### Step 4. How many products cost more than $10.00?

```python
mask_cost=df_chipo.item_price.str.replace("$","").astype(float) >10
```

```python
df_chipo[mask_cost]
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
mask_expensive=df_chipo.item_price.str.replace("$","").astype(float).sort_values(ascending=False)
```

```python
mask_expensive
```

```python
df_chipo.groupby('item_name').sum('quantity').item_price.str.replace("$","").astype(float).sort_values(ascending=False)
```

```python
df_chipo.item_price.str.replace("$","").astype(float).sort_values(ascending=False).head(1)
```

Cambiamos la columna de precios a sin dolares    

```python
prices = [float(value[1 : -1]) for value in df_chipo.item_price]
df_chipo.item_price=prices
```
